# Task 4: Generative Text Model

This notebook demonstrates coherent paragraph generation using GPT-2 based on user prompts.

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
```

```python
def generate_text(prompt: str, max_length: int = 120, temperature: float = 0.8):
    output = generator(
        prompt,
        max_length=max_length,
        num_return_sequences=1,
        do_sample=True,
        temperature=temperature,
        top_k=50,
        top_p=0.95,
    )
    return output[0]["generated_text"]
```

```python
user_prompt = input("Enter a prompt for text generation: ").strip()
if not user_prompt:
    user_prompt = "Artificial intelligence in education will"

generated = generate_text(user_prompt, max_length=140)
print("PROMPT:\n", user_prompt)
print("\nGENERATED TEXT:\n", generated)
```
